```python
# Regular def function
def square(x):
    return x * x

print(square(5))       # 25
```

```python

# Exact same logic as a lambda
square = lambda x: x * x
print(square(5))       # 25 — identical result
```

```python
# 0 arguments
greet     = lambda: "Hello, World!"
print(greet())               # Hello, World!

# 1 argument
double    = lambda x: x * 2
print(double(7))             # 14

# 2 arguments
add       = lambda x, y: x + y
print(add(3, 7))             # 10

# Boolean check
is_even   = lambda n: n % 2 == 0
print(is_even(4), is_even(7))  # True False

# String manipulation
initials  = lambda s: ".".join(w[0] for w in s.split())
print(initials("Amit Dhakad"))  # A.D

# Celsius to Fahrenheit
to_f      = lambda c: (c * 9/5) + 32
print(to_f(100))              # 212.0

# Conditional (ternary) inside lambda
classify  = lambda x: "positive" if x > 0 else "negative" if x < 0 else "zero"
print(classify(-5))          # negative

# Default argument
power     = lambda x, n=2: x ** n
print(power(3), power(3, 3))  # 9, 27

# *args
total     = lambda *args: sum(args)
print(total(1, 2, 3, 4))     # 10

# ML: normalize a value given min and max
normalize = lambda x, mn, mx: (x - mn) / (mx - mn)
print(normalize(50, 0, 100)) # 0.5
```

```python
# Immediately Invoked Lambda — call it without assigning
result = (lambda x, y: x ** 2 + y ** 2)(3, 4)
print(result)   # 25 — Pythagoras!

# Closure: function that returns a lambda (captures outer variable)
def make_multiplier(n):
    return lambda x: x * n   # captures 'n' from outer scope

double   = make_multiplier(2)
triple   = make_multiplier(3)
times_10 = make_multiplier(10)

print(double(5), triple(5), times_10(5))  # 10 15 50

# ML closure: factory that creates a custom scaler
def make_scaler(min_val, max_val):
    return lambda x: (x - min_val) / (max_val - min_val)

age_scaler    = make_scaler(0, 100)
income_scaler = make_scaler(10000, 200000)

print(age_scaler(25))       # 0.25
print(income_scaler(50000)) # 0.2105...

# ⚠️ Classic loop capture bug
fns_bad = [lambda x: x * i for i in range(5)]
print([f(2) for f in fns_bad])  # [8,8,8,8,8] ← all use i=4!

# ✅ Fix: capture value with default argument
fns_good = [lambda x, i=i: x * i for i in range(5)]
print([f(2) for f in fns_good])  # [0,2,4,6,8] ✅
```
