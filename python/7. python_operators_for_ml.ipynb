## What are Operators?

Operators are special symbols that perform operations on values.  
For example: `2 + 3` → here `+` is the operator.

In Machine Learning, we use operators every day to work with numbers, data, and conditions.

---

## 1. Arithmetic Operators

These do basic math.

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `3 + 2` | `5` |
| `-` | Subtraction | `5 - 2` | `3` |
| `*` | Multiplication | `4 * 3` | `12` |
| `/` | Division | `10 / 4` | `2.5` |
| `//` | Floor Division | `10 // 3` | `3` |
| `%` | Modulo (remainder) | `10 % 3` | `1` |
| `**` | Power | `2 ** 3` | `8` |

### Example 1 — Basic math with data

```python
# Imagine a student's exam scores
score1 = 80
score2 = 90
score3 = 70

# Find average score (used in ML to find mean)
total   = score1 + score2 + score3   # addition
average = total / 3                  # division

print(average)   # 80.0
```

### Example 2 — Power operator (used in error calculation)

```python
# Difference between predicted and actual value
actual    = 10
predicted = 8

error    = actual - predicted    # 2
sq_error = error ** 2            # 4  (squaring the error)

print("Error:", error)            # Error: 2
print("Squared Error:", sq_error) # Squared Error: 4
```

> **Why it matters in ML:** Squaring errors is the foundation of MSE (Mean Squared Error), one of the most common ways to measure how wrong a model is.

### Example 3 — Floor division and modulo (splitting data into batches)

```python
total_data = 100
batch_size = 30

full_batches = total_data // batch_size   # 3 full batches
leftover     = total_data % batch_size    # 10 samples left over

print("Full batches:", full_batches)   # 3
print("Leftover samples:", leftover)   # 10
```

> **Why it matters in ML:** During training, we feed data in small groups called "batches". Floor division tells us how many full batches we have.

## 2. Comparison Operators

These compare two values and return `True` or `False`.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal | `5 != 3` | `True` |
| `>` | Greater than | `7 > 4` | `True` |
| `<` | Less than | `3 < 6` | `True` |
| `>=` | Greater or equal | `5 >= 5` | `True` |
| `<=` | Less or equal | `4 <= 4` | `True` |

### Example 1 — Checking a prediction

```python
actual    = 1     # real answer (cat = 1)
predicted = 0.5     # model's answer

if predicted == actual:
    print("Correct prediction!")    # ✅ this prints
else:
    print("Wrong prediction!")
```

### Example 2 — Classifying with a threshold

```python
# Model gives a probability between 0 and 1
probability = 0.75

# If probability is above 0.5, predict class 1 (positive)
if probability > 0.5:
    print("Predicted: Positive")    # Predicted: Positive
else:
    print("Predicted: Negative")
```

> **Why it matters in ML:** This is called "thresholding". Every classifier does this to turn a probability into a final prediction (yes/no, spam/not spam, cat/dog).

### Example 3 — Counting correct predictions (accuracy)

```python
actual_labels    = [1, 0, 1, 1, 0]
predicted_labels = [1, 0, 0, 1, 0]

correct = 0
for a, p in zip(actual_labels, predicted_labels):
    if a == p:         # comparison operator
        correct += 1

print("Correct", correct)
accuracy = correct / len(actual_labels)
print("Accuracy:", accuracy)   # Accuracy: 0.8  → 80%
```

## 3. Logical Operators

These combine multiple conditions together.

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `and` | Both must be True | `True and True` | `True` |
| `or` | At least one True | `True or False` | `True` |
| `not` | Flips True/False | `not True` | `False` |

### Example 1 — `and` operator

```python
age    = 25
income = 30000

# A person qualifies for a loan if BOTH conditions are met
if age >= 18 and income >= 30000:
    print("Loan Approved")     # Loan Approved ✅
else:
    print("Loan Rejected")
```

### Example 2 — `or` operator

```python
has_degree     = True
has_experience = False

# Gets hired if they have a degree OR experience
if has_degree or has_experience:
    print("Hired!")            # Hired! ✅
else:
    print("Not hired.")
```

### Example 3 — `not` operator

```python
is_training = True

# Run evaluation only when NOT training
if not is_training:
    print("Running model evaluation...")   # this prints
```

### Example 4 — Combining all three (data filtering)

```python
age    = 22
score  = 85
passed = False

# Accept if age is right AND score is good AND they passed
if age >= 18 and score >= 70 or passed:
    print("Accepted into program!")   # ✅ prints
else :
    print("Not accepted")
```

> **Why it matters in ML:** Logical operators are used to filter data, set training conditions, and write validation rules.

## 4. Assignment Operators

These store or update values in variables.

| Operator | Example | Same as |
|----------|---------|--------|
| `=` | `x = 5` | assign 5 to x |
| `+=` | `x += 3` | `x = x + 3` |
| `-=` | `x -= 2` | `x = x - 2` |
| `*=` | `x *= 4` | `x = x * 4` |
| `/=` | `x /= 2` | `x = x / 2` |

### Example 1 — Basic assignment

```python
learning_rate = 0.01
epochs        = 100
accuracy      = 0.0

print(learning_rate)   # 0.01
print(epochs)          # 100
```

### Example 2 — `+=` to accumulate loss during training

```python
total_loss = 0.0

# Imagine 5 training steps, each with a loss value
losses = [0.8, 0.6, 0.5, 0.4, 0.3]

for loss in losses:
    total_loss += loss    # add each loss to the total

print("Total loss" , total_loss)
average_loss = total_loss / len(losses)
print("Average Loss:", average_loss)   # 0.52
```

### Example 3 — Updating weights step by step

```python
weight        = 1.0
learning_rate = 0.1
gradient      = 0.5   # how much to adjust

# Gradient descent update — core of ML training

weight -= learning_rate * gradient

print("Updated weight:", weight)   # 0.95
```

> **Why it matters in ML:** Every time a neural network trains, it updates its weights using `-=`. This is called gradient descent.

## 5. Identity Operators

These check if two variables point to the **exact same object** in memory.

| Operator | Meaning |
|----------|---------|
| `is` | Same object |
| `is not` | Not the same object |

### Example 1 — Check if model is ready

```python
model = None   # no model trained yet

if model is None:
    print("Please train the model first!")   # this prints
```

### Example 2 — Check if model exists

```python
model = "RandomForest"   # model is now set

if model is not None:
    print("Model is ready. Making predictions...")   # this prints
```

> **Why it matters in ML:** We often check `if model is None` before using a model, to avoid errors when the model hasn't been trained yet.

## 6. Membership Operators

These check if a value **exists inside** a list, set, or string.

| Operator | Meaning |
|----------|---------|
| `in` | Value is in the collection |
| `not in` | Value is not in the collection |

### Example 1 — Check if a feature exists

```python
available_features = ['age', 'income', 'score', 'gender']

# Check if the feature we need is available
if 'income' in available_features:
    print("Feature found! Ready to use.")    # ✅ prints
```

### Example 2 — Detect missing feature

```python
required  = ['age', 'income', 'credit_score']
available = ['age', 'income']

for feature in required:
    if feature not in available:
        print(f"Missing feature: {feature}")   # Missing feature: credit_score
```

### Example 3 — Check allowed model types

```python
allowed_models = ['linear', 'tree', 'neural_network']
chosen_model   = 'RandomForest'

if chosen_model in allowed_models:
    print(f"Training with: {chosen_model}")    # Training with: tree
else:
    print("Unknown model type!")
```

> **Why it matters in ML:** Before running a pipeline, we always check that required features, column names, and settings are present. `in` and `not in` make this clean and simple.
