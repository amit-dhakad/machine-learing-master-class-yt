### 2. Variables & Naming Conventions
A variable is not a "box" holding a value. It is a **nametag** pointing to a specific address in your computer's RAM. 

**Industry Standard Naming (PEP 8):**
In Python, we strictly use **Snake Case** (all lowercase, words separated by underscores). We do not use Camel Case (`userAge`)—that is reserved for JavaScript and C++.

```python
# ✅ Good naming convention (Snake Case)
student_test_score = 95
learning_rate = 0.001

# ❌ COMMON ERROR 2: SyntaxError (Invalid Names)
# Rule 1: Variables CANNOT start with a number.
# 1st_place_winner = "John"  # This will CRASH

# Rule 2: You CANNOT use Python's reserved keywords (like 'class', 'def', 'if', 'True')
# class = "Math 101"         # This will CRASH
# True = 5                   # This will CRASH

print(f"The student scored {student_test_score} with a learning rate of {learning_rate}.")
```

### 3. The Primitive Data Types
Python is **Dynamically Typed**. You do not have to explicitly tell the CPU what type of data you are creating; it infers it at runtime. 

However, we must constantly audit our data using the `type()` function to ensure our ML pipelines are receiving the correct formats.

**The Core Data Types:**
* `int`: Whole numbers.
* `float`: Decimals (Consume more GPU memory).
* `str`: Text strings.
* `bool`: True or False.
* `complex`: Real + Imaginary numbers (Used in Quantum ML).
* `NoneType`: Represents the intentional absence of a value.

```python
a = 5 + 7j
print(a.real)
type(a)
```

```python
# Let's initialize the core primitives
batch_size = 32                  # int
dropout_rate = 0.5               # float
dataset_path = "data.csv"        # str
is_training = True               # bool
quantum_state = 3 + 5j           # complex (uses 'j' for imaginary)
missing_sensor_data = None       # NoneType (Python's version of Null)

# Audit the memory using type()
print("Batch Size Type:      ", type(batch_size))
print("Dropout Rate Type:    ", type(dropout_rate))
print("Dataset Path Type:    ", type(dataset_path))
print("Missing Data Type:    ", type(missing_sensor_data))

# ❌ COMMON ERROR 3: TypeError
# You cannot perform math between unsupported types (like adding text to a number).
# Uncomment to see the crash:
# print("The score is: " + 95)
# EXPECTED CRASH: TypeError: can only concatenate str (not "int") to str
```

### 4. Type Conversion (Casting)
In the real world, datasets are messy. If a CSV file imports a column of house prices as Text (`str`) instead of Math (`int`), your neural network will fail. We fix this using Casting.

* **Implicit Conversion:** Python automatically upgrades data safely (e.g., `int` + `float` = `float`).
* **Explicit Conversion (Casting):** Forcing data to change forms using `int()`, `float()`, or `str()`.

```python
# --- IMPLICIT CONVERSION ---
integer_val = 10
float_val = 2.5
result = integer_val + float_val 
print(f"Implicit Result: {result} | Type automatically upgraded to: {type(result)}")

# --- EXPLICIT CONVERSION (CASTING) ---
# Scenario: A web scraper pulls a house price, but it arrives as a string of text.
house_price_str = "450000"

# ✅ THE FIX: Cast the string into a mathematical integer
house_price_int = int(house_price_str)
new_price = house_price_int + 10000
print(f"\nSuccessfully performed math after casting: ${new_price:,}")

# ❌ COMMON ERROR 4: ValueError
# You can only cast strings that actually contain clean numbers.
# dirty_data = "450K" # Notice the 'k'
clean_data = int(dirty_data) 
# EXPECTED CRASH: ValueError: invalid literal for int() with base 10: '450k'
```

```python
# Math with Booleans
print("True + True =", True + True)       # 1 + 1 = 2
print("False - True =", False - True)     # 0 - 1 = -1
print("True * 100 =", True * 100)         # 1 * 100 = 100

# Let's simulate evaluating an AI model's predictions
predictions = [True, False, True, True, False]
correct_count = sum(predictions)
print(f"\nThe AI got {correct_count} predictions correct out of {len(predictions)}.")

# ❌ COMMON ERROR 5: ZeroDivisionError
# Because False is exactly 0, dividing by False breaks the fundamental laws of math.
print(True / False) 
# EXPECTED CRASH: ZeroDivisionError: division by zero
```
