### 1. Python Architecture: Indentation & Imports
Unlike C++ or JavaScript, Python **does not use curly braces `{}`** to group code. It uses **Whitespace (Indentation)**. If your indentation is off by a single space, the script crashes.

We also use this section to understand Python's internal vocabulary (`import keyword`) and how to break massive mathematical equations across multiple lines.

```python
# --- 1. INDENTATION ---
if True:
    print("This line belongs to the 'if' block because it is indented.")
    
# ❌ COMMON ERROR 1: IndentationError
# Uncomment the code below to see Python crash:
# if True:
# print("This will crash!") 
# EXPECTED CRASH: IndentationError: expected an indented block

# --- 2. MULTI-LINE STATEMENTS ---
# Method A: The Backslash (\)
calculation_1 = 100 + 200 + 300 + \
                400 + 500

# Method B: Parentheses (Industry Standard for ML Math)
calculation_2 = (
    100 + 200 + 300 + 
    400 + 500
)

# --- 3. IMPORTS & RESERVED WORDS ---
import keyword
print("\nPython Reserved Keywords (DO NOT use these as variable names):")
print(keyword.kwlist)
print(len(keyword.kwlist))


# ❌ COMMON ERROR 2: SyntaxError
# lambda = 5  # 'lambda' is a reserved word. This will crash!
```

### 2. Variable Assignment Mastery
Python allows for highly advanced assignment techniques that save lines of code and execute faster at the hardware level.

```python
# 1. Multiple Assignment (Great for 3D coordinates or RGB values)
x, y, z = 10, 20, 30

# 2. Same Value Assignment (Zeroing out Neural Network weights)
weight_1 = weight_2 = weight_3 = 0.0

# 3. The Python Memory Swap
# In other languages, swapping requires a temporary third variable. In Python:
a = "Apples"
b = "Bananas"
print(f"Before Swap: a={a}, b={b}")

a, b = b, a  # SWAPPED INSTANTLY!
print(f"After Swap:  a={a}, b={b}")
```

### 3. Interactive Data & The Casting Trap
We use the `input()` function to stream live user data into our scripts. 

**⚠️ THE GOLDEN RULE:** The `input()` function **ALWAYS** returns a String (`str`). Even if the user types `50`, Python sees it as text `"50"`. You must explicitly cast it before doing math.

```python
user_name = input("Enter your name: ")
user_age_str = input("Enter your age: ")

print(f"\nWelcome, {user_name}! System sees your age as type: {type(user_age_str)}")

# ❌ COMMON ERROR 3: TypeError
# future_age = user_age_str + 5  # CRASH! Cannot add int to str.

# ✅ THE FIX: Explicit Casting
user_age_int = int(user_age_str)
future_age = user_age_int + 5
print(f"In 5 years, you will be {future_age}.")
```

### 4. Decision Engines (`if`, `elif`, `else`)
We use control flow to make our scripts evaluate conditions top-down. 
**Nested Logic** is when we place an `if` statement *inside* another `if` statement to drill down into complex decisions (this is the foundation of Decision Tree algorithms).

```python
print("--- AI Loan Approval System ---")
income = int(input("Enter your yearly income: "))
credit = int(input("Enter credit score (300-850): "))

# 1. Outer If (Primary condition)
if income >= 100000:
    print("Income tier: High. Evaluating credit...")
    
    # 2. Nested Logic
    if credit >= 750:
        print("✅ Premium Loan Approved!")
    else:
        print("❌ Denied: Credit too low for Premium tier.")

# 3. Elif (Secondary condition)
elif income >= 50000:
    print("Income tier: Moderate. Evaluating risk...")
    if credit >= 700:
        print("✅ Standard Loan Approved.")
    else:
        print("❌ Denied: Need 700+ credit for this tier.")

# 4. Else (Absolute fallback)
else:
    print("❌ Denied: Income below minimum threshold.")
```

```python
age = int(input("enter your age"))

if age < 15 :
    print(f"first if your age is less than {age}")
elif age < 20 :
    print(f" first elif your age is less than {age}")
elif age < 30: 
    print(f"second elif your age is less than {age}")
else : 
    print(f"else your age is less than {age}")
    
```

### 5. The Iteration Engine (`while` loops & `break`)
If we need to run an operation 10,000 times, we don't copy and paste code—we write a loop. 

A `while` loop runs continuously as long as a condition is `True`. If you forget to update the condition, it creates an **Infinite Loop** that will crash your CPU. We use the `break` keyword as an emergency stop.

```python
# --- 1. Standard While Loop ---
battery = 100
while battery > 80:
    print(f"System running... Battery at {battery}%")
    battery -= 5 # CRUCIAL: Updating the variable to prevent an infinite loop

print("Test Complete.\n")

# --- 2. Infinite Loop & Break Logic ---
security_attempts = 0

while True: # This loop runs forever by default
    security_attempts += 1
    print(f"Unauthorized attempt {security_attempts}...")
    
    if security_attempts >= 3:
        print("🚨 MAXIMUM ATTEMPTS. TRIGGERING LOCKOUT.")
        break # Shatters the loop instantly!
```

### 6. Data Streaming (`for` loops & `continue`)
A `for` loop executes code a *specific* number of times. In ML, we use this to train models over "Epochs" using the `range(start, stop)` function.

If we encounter a broken row of data while looping, we don't want to `break` the whole loop. We use `continue` to skip the bad data and move to the next iteration.

```python
# --- 1. For Loops with Range ---
# range(1, 6) generates: 1, 2, 3, 4, 5
print("--- Training AI Model ---")
for epoch in range(1, 6):
    print(f"Epoch {epoch}/5 complete.")

print("\n--- Processing Sensor Data ---")
# --- 2. The Continue Keyword ---
for reading in range(1, 6):
    if reading == 3:
        print("⚠️ Glitch at reading 3. Skipping!")
        continue # Skips to the next iteration. The print below is ignored for '3'.
        
    print(f"Processing reading: {reading}")
```

### 7. NLP Foundations: Strings & Formatting
The final step in our logic engine is manipulating text. 
* **Methods:** `.strip()` removes outer whitespace, `.upper()` capitalizes, `.replace()` swaps text.
* **Slicing:** Ripping out chunks of text using `[start : stop]`.
* **Formatting:** Using f-strings to lock decimal precision (`.2f`).

```python
raw_text = "   Machine Learning is the Future!   "

# 1. String Methods
clean_text = raw_text.strip()
print(f"Stripped: '{clean_text}'")
print(f"Uppercase: '{clean_text.upper()}'")
print(f"Replaced: '{clean_text.replace('Future', 'Present')}'")

# 2. Slicing [start:stop]
# Extract "Machine" (indices 0 through 7)
print(f"Sliced Word: {clean_text[0:7]}")

# 3. Output Formatting (Decimals and Currency)
ai_confidence = 0.987654321
revenue = 4500000.5

print(f"\nRaw Confidence: {ai_confidence}")
print(f"Formatted Confidence: {ai_confidence:.2f}") # Locks to 2 decimals
print(f"Total Revenue: ${revenue:,.3f}") # Adds commas and locks decimals


# --- CAPSTONE: COMBINING LOOPS AND STRINGS ---
# We can loop through a string to clean out bad characters!
dirty_data = "M1a2c3h4i5n6e"
extracted_word = ""

for char in dirty_data:
    if char.isalpha(): # If the character is a letter...
        extracted_word += char # Add it to our new string

print(f"\nCapstone Cleaned Data: {extracted_word}")
```
